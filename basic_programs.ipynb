9. Area of circle

```python
from math import pi
# the number pi is stored in math module
# you can either fetch it from math or write it manually 

print('-- Area of Circle --')

radius = int(input('Enter Radius of the Circle: '))

area = round(2*pi*radius*radius, 2)
print(f'Area of circle with radius:{radius} is {area}')
```

11. Simple Interest

```python
print('-- Simple Interest --')

p = int(input('Enter priciple amount(only digits): '))
r = int(input('Enter interest rate(0-100): '))
t = int(input('Enter time period(only digits): '))

#round() funtion rounds up a number upto given decimal places
#in this example upto 2 decimal places
interest = round((p*r*t)/100, 2)

print(f'The total interest on principle amount {p} with an interest rate of {r}%  for {t} years is {interest}')
```

12. Temperature(°C to °F)

```python
print('-- temperature (°C to °F) --')

c = int(input('Enter temperature in Celcius: '))
f = (c*9/5) + 32

f = round(f, 2)
print(f'{c}°C = {f}°F')
```

13. Temperature(°F to °C)

```python
print('-- temperature (°F to °C) --')

f = int(input('Enter temperature in Fahrenheit: '))
c = (f-32) * 5/9

c = round(c, 2)
print(f'{f}°F = {c}°C')
```

14. Days into Weeks, months & Years

```python
print('-- Days into Weeks, months & Years --')

days = int(input('Enter number of days: '))

# Use int() to find exact number of weeks, moths & years
weeks = int(days/7, 0)
months = int(days/30, 0)
years = int(days/365, 0)

#use % to find number of remaining days
w_rem = days % 7
m_rem = days % 30
y_rem = days % 365

print(f'{days} days = ')
print(f'\t{weeks} weeks and {w_rem} days.')
print(f'\t{months} months and {m_rem} days.')
print(f'\t{years} years and {y_rem} days.')
```
